```
import matplotlib.pyplot as plt; 
plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt

objects = ('CDS1', 'CDS2', 'CDS3')
y_pos = np.arange(len(objects))
performance = [25.79,26.78,21.45]
plt.figure(figsize=(4,4))
for a,b in zip(y_pos, performance):
    plt.text(a-0.3, b+0.04, str(b))

plt.bar(y_pos, performance, align='center', alpha=0.5)
plt.xticks(y_pos, objects)
plt.ylabel('Average GFLOPS')
plt.xlabel('Dataset')
plt.title('Single-node CPU Performance')

plt.show()
```
