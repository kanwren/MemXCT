```
#memxct multinode scaling (figure 11) (demo data, but we will need 4 of these)
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# Data
#fixed
df=pd.DataFrame({'x': [1,2,4], #nodes
                 'Total': [3.1508387,4.2112689,6.465829], 
                 'Ap': [2.03283229,1.89022054,2.5159455], 
                 'C': [0.07942261,0.09407656,0.16610658], 
                 'R': [0.58244757,1.30022732,2.07288389] })

# multiple line plot
plt.plot( 'x', 'Total', data=df, marker='o', markerfacecolor='blue', markersize=12, color='blue', linewidth=4)
plt.plot( 'x', 'Ap', data=df, marker='o', markerfacecolor='olive', markersize=12, color='olive', linewidth=4)
plt.plot( 'x', 'C', data=df, marker='o', markerfacecolor='red', markersize=12, color='red', linewidth=4)
plt.plot( 'x', 'R', data=df, marker='o', markerfacecolor='green', markersize=12, color='green', linewidth=4)

plt.title("Strong Scaling (CPU, CDS1)")
plt.xlabel("Number of nodes")
#plt.ylim(ymin=0)
plt.yscale('log')
plt.ylabel("Solution time (s)")
plt.legend()

```
